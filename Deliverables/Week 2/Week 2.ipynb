**2. Write a program to create a recursive function to calculate the sum of numbers from 0 to 15.**

```python
def sum_(n):
    if n == 0:
        return 0
    else:
        return n + sum_(n-1)
    
print('The sum of 0 to 15 is:',sum_(15))
```

**3. Using Matplotlib, complete the following tasks using the data contained here: https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv:**
    
    a. Exercise 1: Read Total profit of all months and show it using a line plot.
    b. Exercise 2: Read Bathing soap facewash of all months and display it using the Subplot.**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
comp = pd.read_csv('company_sales_data.csv')
comp.head()
```

```python
comp = comp.set_index('month_number', drop = True)
comp['total_profit'].plot()
plt.xticks(range(1,13))
plt.ylabel('total_profit')
plt.title('1. Total profit of all the months')
plt.show()
```

```python
fig, ax = plt.subplots(1,2,figsize=[10,5])
ax[0].plot(comp['facewash'])
ax[0].set_title('facewash')
ax[0].set_xticks(range(1,13))
ax[1].set_title('bathingsoap')
ax[1].plot(comp['bathingsoap'])
ax[1].set_xticks(range(1,13))
fig.suptitle('2. Subplot showing facewash and bathingsoap for all months')
plt.show()
```
